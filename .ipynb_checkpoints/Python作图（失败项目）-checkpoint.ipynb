# 准备工作

```python
#建立连接
def acquire_mysql_data(sql , host = 'localhost' ,database = 'practise' ,user = 'root' ,password = '1987' ):
    import pymysql 
    import numpy as np 
    import pandas as pd 
    conn = pymysql.connect(host,user,password, database ,charset = 'utf8')
    try:
        result = pd.read_sql(sql,conn)
        return result 
    except Exception as e :
        print (e)
```

```python

#import 必要的包
import numpy as np 
import pandas as pd 
from pandas import DataFrame
import matplotlib.pyplot as plt 
import matplotlib
myfont = matplotlib.font_manager.FontProperties(fname='C:/Windows/Fonts/msyh.ttf')  
%matplotlib inline 
```

# 查看总业绩
* 偷懒一下，直接通过sql取数

```python
sql = '''select sum(订货额),sum(销售额),round((sum(订货额)-sum(销售额))*100/sum(订货额),1) as 退货率
from staff_detail'''
result = acquire_mysql_data(sql)
order_amount,sale_amount,refund_rate= result.values[0,0],result.values[0,1],result.values[0,2]

```

```python
print ('总订货额为%d元，总销售额为%d元，退货率为%s%%'%(order_amount,sale_amount,refund_rate))
```

# 每日业绩

```python
sql = '''
select 日期 as sale_date,
sum(销售额) as order_amount ,sum(订货额) as refund_amount ,
round(sum(退货额)/sum(订货额),3) as refund_rate
from staff_detail
group by 日期'''
result = acquire_mysql_data(sql)
```

```python
result.head()
```

```python
#result['refund_new'] = np.where(result['refund_rate']>= 0.15,0.15,result['refund_rate'])
#result[['refund_new']].plot()

x=result['sale_date']
y1 = result['order_amount']
y2 = result['refund_amount']
plt.plot(x,y1)
plt.plot(x,y2)
plt.title('11月的订购与退货',fontproperties =myfont )
plt.xlabel('日期' ,fontproperties =myfont)
```

```python
sql = '''select 日期 as sale_date, 项目 as prj , sum(`订货额`) order_amount
from staff_detail 
group by 日期, 项目'''
result = acquire_mysql_data(sql)
```

```python
result = result.pivot_table(values = 'order_amount' , index = 'sale_date' , columns = 'prj' , aggfunc ='sum')
```

```python
p = kk.line_tidy(result, area=True, stacked=True, )
p.set_size(width=1000) 
p.set_toolbox(restore=True, data_zoom=True)
p.set_toolbox(magic_type=['line','bar'],restore=True)
```

```python
sql = '''select 日期 as sale_date,  sum(`订货额`) order_amount
from staff_detail 
group by 日期'''
result = acquire_mysql_data(sql)
```

```python
p = kk.waterfall(result['order_amount'],annotate="outside", up_name="up")
p.set_size(width=1200) 
```

```python
sql = '''select 姓名 as name , sum(分配数据量) data_sum, sum(通时) call_sum , sum(销售额) sale_amount
from staff_detail
where 姓名 !=''
group by 姓名'''
result = acquire_mysql_data(sql)
result.head()
```

```python
p = kk.scatter(result,'data_sum','call_sum',s = 'sale_amount', c =  'sale_amount')
p.set_size(width=1000, height=500)
p.set_tooltip_format(['name','data_sum','call_sum','sale_amount'])
p.set_theme('dark')
p.set_toolbox(save_format='png',restore=True,data_zoom=True)
p.set_legend(orient='vertical',x_pos='-1%',y_pos='-3%')
p.set_title('data call sale',x_pos='center',y_pos='-3%')
```
